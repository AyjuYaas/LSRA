```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel("./E Commerce Dataset.xlsx", sheet_name="E Comm")
display(data.columns)
data['Churn'] = data['Churn'].replace({0:"No", 1:"Yes"})
```

```python
# Creating Tenure Groups
bins = [0, 6, 12, 18, 24, np.inf]
labels = ['0-6', '6-12', '12-18', '18-24', '24+']
data['TenureRange'] = pd.cut(data['Tenure'], bins = bins, labels = labels, right = False, include_lowest = True)

plt.figure(figsize=(6,4))
ax = sns.countplot(data = data, x = 'TenureRange', hue = 'Churn')

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha = 'center', va = 'baseline', 
                    xytext = (0, 3), weight = 'bold',
                    textcoords = 'offset points')

plt.xlabel('Tenure(months)')
plt.ylabel('Frequency')
plt.show()
```
